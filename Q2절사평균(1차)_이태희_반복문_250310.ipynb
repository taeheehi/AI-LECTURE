```python
import pandas
```

```python
# 깃허브의 www.github.com/hyowakn > python_lecture > kopo_customerdata.csv
csFilePath = 'https://raw.githubusercontent.com/hyokwan/python-lecture/refs/heads/master/dataset/kopo_customerdata.csv'
customerData = pandas.read_csv(csFilePath)

# Step1. 리스트 형태로 변경한다.
customerTotalList = list(customerData.TOTAL_AMOUNT)  # 리스트 형태로 변경

# Step2. 최소값, 최대값 찾기
# 최소값
minValue = min(customerTotalList)
minValue # 0

# 최대값
maxValue = max(customerTotalList)
maxValue # 3103500


# Step3. 리스트의 최소값과 최대값을 제거한다.
customerTotalList.remove(minValue)
customerTotalList.remove(maxValue)


# Step4. 최대 최소값 다시 확인작업
# minValue = min(customerTotalList)
# minValue # 그대로 0... 안바뀌었나..? 일단 스킵

# maxValue = max(customerTotalList)
# maxValue # 2441300 : 변경완료


# Step5. 리스트의 평균을 계산한다. (전체합 / 전체개수)
sumList = sum(customerTotalList)
cntList = len(customerTotalList)
answer = sumList / cntList
answer    # 35218.95312635417
```
